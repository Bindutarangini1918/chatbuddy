```python
import fitz  # PyMuPDF

def extract_text_from_pdf(file_path):
    text = ""
    with fitz.open(file_path) as pdf:
        for page in pdf:
            text += page.get_text()
    return text

pdf_file = "/content/drive/MyDrive/cyber security audit.pdf"  # Update with your PDF file path
extract_text_from_pdf(pdf_file)
```

```python
def summarize_report(report_text):
    prompt = f"Summarize the following cybersecurity audit report: {report_text}"
    response = genai.GenerativeModel("gemini-1.5-flash").generate_content(prompt)
    return response.text.strip()
```

```python
def answer_question(report_text, user_query):
    prompt = f"Answer the following question based on the audit report: {user_query}\nReport: {report_text}"
    response = genai.GenerativeModel("gemini-1.5-flash").generate_content(prompt)
    return response.text.strip()
```
